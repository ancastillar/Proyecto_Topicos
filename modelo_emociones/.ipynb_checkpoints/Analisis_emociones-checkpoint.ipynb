```python
def emotion_count(text,vocab):
    # Separamos las pal√°bras por espacios.
    words=text.split(" ")
    # Creamos un diccionario donde se guardar√°n los conteos por cada emoci√≥n.
    counts={i:0 for i in list(vocab.keys())}
    # Creamos un diccionario donde se guardar√°n las palabras coincidentes con cada l√©xico.
    words_per_emo={i:[] for i in list(vocab.keys())}
    # Iteramos para cada una de las pal√°bras dentro del texto.
    for word in words:
        # Iteramos para cada una de las emociones del l√©xico.
        for emo in vocab:
            # Eval√∫amos si la palabra est√° dentro del l√©xico de cada emoci√≥n
            if word in vocab[emo]:
                # Si la palabra est√° en el l√©xico de la emoci√≥n, sumamos 1 al conteo acumulado.
                counts[emo]+=1
                # Tambi√©n agregamos la palabra coincidente.
                words_per_emo[emo].append(word)
    return counts, words_per_emo
```

```python
###Cargue de datos
data = pd.read_csv("text_emotion.csv")
data0=pd.read_csv(ruta+"train.txt", sep=';',names=["content","sentiment"])
data=data[['content','sentiment']]
data0=data0[['content','sentiment']]
data=pd.concat([data, data0],axis=0)
```

**Para eliminar algunos errores comunes de escritura**

```python
misspell_data = pd.read_csv(ruta+"spell.txt",sep=":",names=["correction","misspell"])
misspell_data.misspell = misspell_data.misspell.str.strip()
misspell_data.misspell = misspell_data.misspell.str.split(" ")
misspell_data = misspell_data.explode("misspell").reset_index(drop=True)
misspell_data.drop_duplicates("misspell",inplace=True)
miss_corr = dict(zip(misspell_data.misspell, misspell_data.correction))

#Visualizaci√≤n del diccionario
{v:miss_corr[v] for v in [list(miss_corr.keys())[k] for k in range(20)]}
```

**Funci√≥n para quitar las contracciones**

```python
def misspelled_correction(val):
    for x in val.split(): 
        if x in miss_corr.keys(): 
            val = val.replace(x, miss_corr[x]) 
    return val
data["clean_content"] = data.content.apply(lambda x : misspelled_correction(x))

#Contracciones comunes en ingl√©s
contractions = pd.read_csv(ruta+"datasets_697638_1220322_contractions.csv")
cont_dic = dict(zip(contractions.Contraction, contractions.Meaning))

def cont_to_meaning(val): 
  
    for x in val.split(): 
        if x in cont_dic.keys(): 
            val = val.replace(x, cont_dic[x]) 
    return val
data.clean_content = data.clean_content.apply(lambda x : cont_to_meaning(x))
data["clean_content"]=data.content.apply(lambda x : p.clean(x))
```

**Al trabajar con datos de Twitter es necesario limpiar, emojis, puntuaciones**

```python
#Puntuaction and emojis
def punctuation(val): 
  
    punctuations = '''()-[]{};:'"\,<>.\~‚Ç¨¬¨+*/@#$%^?&_~|!¬°'''
  
    for x in val.lower(): 
        if x in punctuations: 
            val = val.replace(x, " ") 
    return val

data.clean_content = data.clean_content.apply(lambda x : ' '.join(punctuation(emoji.demojize(x)).split()))

##Peque√±a limpieza del texto

def clean_text(val):
    val = misspelled_correction(val)
    val = cont_to_meaning(val)
    val = p.clean(val)
    val = ' '.join(punctuation(emoji.demojize(val)).split())
    
    return val
    
data = data[data.clean_content != ""]
data.clean_content = data.clean_content.apply(lambda x: x.lower())
```

```python
#MODELO
#Mapeamos cada una de las emociones
sent_to_id  = {"empty":0, "sadness":1,"enthusiasm":2,"neutral":3,"worry":4,
                        "surprise":5,"love":6,"fun":7,"hate":8,"happiness":9,"boredom":10,"relief":11,"anger":12, "fear":13, "joy":14}
data["sentiment_id"] = data['sentiment'].map(sent_to_id)
```

**Vectorizacion del texto**
----------------------------------------------------

```python
X = data.clean_content
y = data.sentiment_id

########################vectorizacion de las frases

# Maximo numero de palabra para emplear
MAX_NB_WORDS = 50000

MAX_SEQUENCE_LENGTH =  100 # Cantidad m√°xima de secuencias

token = text.Tokenizer(num_words=MAX_NB_WORDS, filters='!"#$%&()*+,-./:;<=>?@[\]^_`{|}~', lower=True)

token.fit_on_texts(X.values)

word_idx = token.word_index

print('Se encuentran %s tokens √∫nicas' % len(word_idx))

X_pad = sequence.pad_sequences(token.texts_to_sequences(X), maxlen= MAX_SEQUENCE_LENGTH )

print('Dimension del tensor resultante:', X_pad.shape)


##Convertir los targets (ya categ√≥ricos) a numeros 

y = pd.get_dummies(y.values)
print('Dimension del tensor target:', y.shape)
```

**Partimos los datos en entrenamiento y en conjunto de prueba**

```python
X_train, X_test, y_train, y_test = train_test_split(X_pad, y, test_size = 0.15, random_state = 42, shuffle= True)
print("Entrenamiento", X_train.shape,y_train.shape)
print("Prueba", X_test.shape,y_test.shape)
```

**Dise√±o de la arquitectura de la red**
------------------------------------------------

```python
embed_dim = 100 
lstm_out = 200
batch_size = 80
epoch = 15
#################Arquitectura

model = Sequential()
model.add(Embedding(len(word_idx) +1 , embed_dim, input_length = X_train.shape[1])) #Esta capa crea una representaci√≤n vectorial de dimension embed_dim para cada palabra
model.add(SpatialDropout1D(0.3)) #Dropout para modelos NLP
model.add(SpatialDropout1D(0.45)) 
model.add(LSTM(lstm_out, dropout=0.4, recurrent_dropout=0.4, activity_regularizer=regularizers.l1(0.5)))
model.add(Dense(units=64, activation='relu',activity_regularizer=regularizers.l2(0.01)))
model.add(Dense(15, activation='softmax'))# 15 CLASES 
#Modelo


callback = EarlyStopping(monitor='loss', patience=3)
model.compile(loss = "categorical_crossentropy", optimizer='adam',metrics = ['accuracy'])
print(model.summary())
```

```python
%%time
model.fit(X_train, y_train, epochs = epoch, batch_size=batch_size, validation_data=(X_test, y_test), callbacks= [callback])

model.save(ruta+'modelo_emociones/model_emociones.h5')
```
